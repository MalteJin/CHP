## General motivation

Micro grids hosting renewable, inverter based generation (wind, photovoltaic) and conventional, synchronous generator based generation units at same time are interesting study objects. Practically all effects of large-scale integration of renewable, inverter based generation into a larger, nation-wide grid can be studied in micro grids setups. Grid stability enhancements methods can be evaluated in micro grids setups and are also applicable on national grid level. Thus, stability considerations for micro grids give valuable knowledge also for larger systems.

##  Specific motivation

Combined heat and power units (CHPs) in  Germany  are required to prove grid code compliance since 2015. Grid code compliance is certified within the framework defined by FGW guidelines TR3, TR4 and TR8. These FGW guidelines are developed in working groups since 1992. Participants of working groups are measuring institutions, advisory firms, network operators, certifiers and producers of wind turbines and photovoltaic systems as well as more recently, manufacturers of CHPs.

The generator technology of CHPs above 30 kW output power usually features a directly grid-connected synchronous generator <a name="ref-1"/>[(Rosa, 2013)](#cite-Rosa2013), p. 32. Unlike inverter based generator technology, as commonly found in wind turbine or photovoltaic systems, the operational performance of directly grid-connected synchronous generators can be assessed in a general way with only a few data sheet values. In this work certification relevant characteristics of CHPs with directly grid-connected synchronous generator are derived based on literature knowledge and data sheet values.

## Derived quantities

Starting with the data sheet values and using the calculation scheme of 
<a name="ref-3"/>[(Crastan)](#cite-Crastan), p. 259 f., the rotor voltage, rotor angle and further values are calculated for the nominal operating point.

```python
Sn =Sn_widg.value
Un =Un_widg.value
fn =fn_widg.value
Nn =Nn_widg.value
If0=If0_widg.value
Rf =Rf_widg.value
xd =xd_widg.value
xq =xq_widg.value
r  =r_widg.value
cosPhiN=cosPhiN_widg.value

If0 =If0/1e3 #Convert to kA
UnLg =Un/np.sqrt(3) #Line to ground nominal voltage
Zn=Un**2/Sn
Xd= xd*Zn
Xq= xq*Zn
R =r*Zn
In=Sn/(np.sqrt(3)*Un)
phiN=np.arccos(abs(cosPhiN))
InCpx = In*np.exp(1j*phiN*np.sign(cosPhiN))
#Operating point values (set to nominal values)
U0LgCpx=Un/np.sqrt(3) #Note division by sqrt(3)
U0Lg=np.abs(U0LgCpx)

Icpx =InCpx
I=np.abs(Icpx)
cosPhi=cosPhiN
phi=np.arccos(abs(cosPhi))
w=2*np.pi*fn
wm =Nn/60 *2*np.pi
n =w/wm #Relative or pu rotational speed (Crst 6.13)
EptCpx=U0LgCpx+R*Icpx+1j*Xq*Icpx
Ept=np.abs(EptCpx)
dlt=np.angle(EptCpx)

EpCpx=Ept+(Xd-Xq)*I*np.sin(phi+dlt)
Ep=np.abs(EpCpx)
Ud=U0Lg*np.sin(dlt)
Uq=U0Lg*np.cos(dlt)
Id=I*np.sin(dlt+phi)
Iq=I*np.cos(dlt+phi)

Xhfd= UnLg*np.sqrt(2)/If0
IfDc =Ep*np.sqrt(2)/Xhfd
UfDc =IfDc *Rf

Scpx = 3*U0LgCpx*np.conj(Icpx)
Ploss= 3* np.real(R/3*Icpx**2) #Stator losses  
Pei=np.real(Scpx)+Ploss
M=Pei/wm

display(Latex('Calculated nominal values:'))
display(Latex('Nominal current $I_n$: In= %.4g kA' %In))
display(Latex('Base impedance $Z_n$: Zn= %g $\Omega$' %Zn))
display(Latex('Reactance $X_d$: Xd= %.4f $\Omega$' %Xd))
display(Latex('Reactance $X_q$: Xq= %.4f $\Omega$' %Xq))

display(Latex('$E_{pt}$: Ept= %.4f kV (%.2f   pu) '%(Ept,Ept/UnLg)))
display(Latex('Load angle $\delta$: dlt= %.2f $^{\circ}$' %(dlt*180/np.pi)))
display(Latex('(Synchronous generated) internal voltage $E_p$: Ep= %.4f kV (%.2f   pu)'%(Ep,Ep/UnLg)))
display(Latex(
    'Amplitude of internal voltage$\hat{E}_p$: EpInst= %.4f kV (%.2f   pu)'
        %(Ep*np.sqrt(2),Ep*np.sqrt(2)/UnLg)))
display(Latex(
        'd-axis component of stator voltage: $U_d$: Ud= %.4f kV (%.2f   pu)'
         %(Ud,Ud/UnLg)))
display(Latex(
        'q-axis component of stator voltage: $U_q$: Uq= %.4f kV (%.2f   pu)'
         %(Uq,Uq/UnLg)))
display(Latex('d-axis component of stator current: $I_d$: Id= %.4f kA (%.2f   pu)'
              %(Id,Id/In)))
display(Latex('d-axis component of stator current: $I_q$: Iq= %.4f kA (%.2f   pu)'
              %(Iq,Iq/In)))
display(Latex('Coupling reactance: $X_{hfd}$: Xhfd= %.4f $\Omega$' %Xhfd))
display(Latex('Excitation DC current: $I_f$: Ifd= %.4f A (DC)'%(IfDc*1e3)))
display(Latex('Excitation DC voltage: $U_f$: Ifd= %.4f V (DC)'%(UfDc*1e3)))
display(Latex('Excitation power losses: $P_f$: Pf= %.4f kW (DC)'
              %(UfDc*IfDc*1e3))) #*3?
display(Latex(
        'Internal electrical power: $P_{ei}$: Pei= %.4f MW (%.2f   pu)'
         %(Pei,Pei/Sn)))
display(Latex('Internal torque: $M$: M= %.4f kNm '%(M*1e3)))
```

##  Phasor diagram

With the values for the variables $ U, I, cos \ varphi, X_d $ and $ I_n $ the phasor diagram can be drawn for a turbo generator (assuming  $x_d=x_q$). The variables $ U, I, cos \varphi $ are interpreted as input parameters for the operating state. Using the sliders widgets the phasor diagram can be interactively adjusted according to the respective operation point.

```python
from pylab import rcParams
rcParams['figure.figsize'] = 10, 10 #Set plot box size. Adapt if necessary.

def drawPhasors(U,I,cosPhi):

   phi=np.arccos(abs(cosPhi))
   Icpx = I*np.exp(1j*phi*np.sign(cosPhi))
   UlgCpx=U/sqrt(3)  

   EptCpx=UlgCpx+R*Icpx+1j*Xq*Icpx
   Ept=np.abs(EptCpx)
   dlt=np.angle(EptCpx)
   Ep=Ept+(Xd-Xq)*I*np.sin(phi+dlt)
   EpCpx= Ep*np.exp(1j*dlt)
  
   IcpxScale=5 #Scaling factor for better viewability of I phasor.
   
   Uvec   = PositionVectorClass(np.array([0,0]),UlgCpx)
   EpVec  = PositionVectorClass(np.array([0,0]),EpCpx)
   Ivec   = PositionVectorClass(np.array([0,0]),Icpx*IcpxScale)
   IXdVec = PositionVectorClass(Uvec.b,Icpx*1j*Xd )
   
   Uvec.draw(r'$\underline{U}$',arrowColor='black' ,textColor='black',
             xTextOffset=0 ,yTextOffset= 5)
   EpVec.draw(r'$\underline{E}_p$',arrowColor='red',textColor='red',
              xTextOffset=-30 ,yTextOffset= 0)
   Ivec.draw( str(IcpxScale)+r'$\underline{I}$',arrowColor='blue',
             textColor='blue',xTextOffset=0 ,yTextOffset= -25)
   IXdVec.draw(r'$\underline{I}X_d$',arrowColor='green' ,textColor='green'
               ,xTextOffset=15,yTextOffset= 0)
  
   Uvec.angleDraw(EpVec,r'$\delta$')
   Uvec.angleDraw(Ivec,r'$\varphi$')
    
   plt.ylim(-8,8) #Adapt if necessary
   plt.xlim(-8,8)
  
   plt.grid()    
   ax=plt.gca()
   ax.set_xlabel('Re')
   ax.set_ylabel('Im')
   font = {'weight' : 'normal','size':16}
   rc('font', **font) 
   #http://stackoverflow.com/questions/3899980/
   #how-to-change-the-font-size-on-a-matplotlib-plot
   
   textLine1 =('$E_p=%.4f\enspace\mathrm{kV}\enspace(%.2f\mathrm{pu})$'
                    %(np.abs(EpCpx),np.abs(EpCpx)/(Un/sqrt(3))) ) 
   
   textLine2 =('$\delta=%.2f^{\circ}$'
                    %np.angle(EpCpx,deg=True)) 
   
   textLines =  "%s\n%s" %(textLine1,textLine2)

   props = dict(boxstyle='round', facecolor='white', alpha=0.5)
   ax.text(0.05, 0.95, textLines, transform=plt.gca().transAxes, fontsize=18,
   verticalalignment='top', bbox=props, color='red')
   #Add boxed text: http://matplotlib.org/users/recipes.html
  
sldSteps=100
U_Slider = widgets.FloatSlider(min= 0, max= Un,step=Un/sldSteps, 
                               value=round(Un,3),description="$U$ [kV]")
I_Slider = widgets.FloatSlider(min= -In, max=In,step=2*In/sldSteps, 
                               value=round(In,3),description="$I$ [kA]")                                             
cosPhi_Slider = widgets.FloatSlider(min=-1, max=1,step=2/sldSteps, 
                                   value= cosPhiN,description="$cos\\varphi$")
widgets.interactive(lambda a,b,c: drawPhasors(a,b,c),
                    a=U_Slider,b=I_Slider,c=cosPhi_Slider)
```

#  Power chart

The power chart ("PQ-chart") of the round rotor machine describes the limits of reactive power control range in the stationary case. The boundaries of the power chart are given by different technical limits. The individual limits are discussed in this section.

## Stator current limit

The thermal limit of the stator current at nominal current is defined by nominal power and voltage values. The limits in the power chart correspond to a circle around the origin with radius equal to the nominal power. In explicit form this circle can be described by two root functions.

```python
QmaxStatAFun=lambda P:    np.sqrt(Sn**2-P**2)
QmaxStatBFun=lambda P: -1*np.sqrt(Sn**2-P**2)
```

## Rotor current limit

The derivation of the formulas for the thermal limit of the rotor current at nominal internal voltage starts with the formulas from <a name="ref-4"/>[(Crastan)](#cite-Crastan), p. 274. For the active and reactive power output of the ideal round rotor machine the formulas apply to:

```python
Psym,Qsym,uSym,QdSym,EpSym,UnSym,dltSym=sp.symbols(
                                           r"P,Q,u,Q_d,E_p,U_n,\delta")
eq0= uSym*QdSym*EpSym/UnSym *sp.sin(dltSym) -Psym  #  P(dlt) term (Crst 6.94)
eq1= uSym*QdSym*(EpSym/UnSym *sp.cos(dltSym) -uSym)-Qsym # Q(dlt) term 
display(sp.Eq(Psym,eq0+Psym))
display(sp.Eq(Qsym,eq1+Qsym))
```

In these expressions it is $\delta$  the rotor angle and $q_d$ the so called characteristic reactive power with  $Q_d=\frac{3U^2_r}{X_d}$. The per unit stator voltage $u$ is equal to the nominal voltage, this means assumed to 1 pu.

```python
Qd=3*Un**2/Xd #(CRST 6.90)
u=1
```

The above expressions for $P(\delta)$ and $Q(\delta)$ can be both solved for $\delta$ first, and the be inserted into each other in order to replace $\delta$. The result is two terms for $P(Q)$ and $Q(P)$.

```python
eq2=sp.solve(eq1,dltSym) #Solve Q(dlt) for load angle delta
eq3a=eq0.subs(dltSym,eq2[0]) #Insert first solution for dlt in P(dlt) 
eq3b=eq0.subs(dltSym,eq2[1]) #Insert second solution for dlt in P(dlt)
display(sp.Eq(Psym(Qsym),eq3a+Psym))#Show 1. term for P(Q) with elimiminated delta
display(sp.Eq(Psym(Qsym),eq3b+Psym))#Show 2. term for P(Q) with elimiminated delta

eq4=sp.solve(eq0,dltSym) #Solve P(dlt) for load angle delta
eq5a=eq1.subs(dltSym,eq4[0]) #Insert first solution for dlt in Q(dlt) 
eq5b=eq1.subs(dltSym,eq4[1]) #Insert second solution for dlt in Q(dlt)
display(sp.Eq(Qsym(Psym),eq5a+Qsym))#Show 1. term for Q(P) with elimiminated delta
display(sp.Eq(Qsym(Psym),eq5b+Qsym))#Shwo 2. term for Q(P) with elimiminated delta
```

In <a name="ref-5"/>[(Crastan)](#cite-Crastan), p. 274 f. the characteristics of similar triangles are used for the derivation of the power chart. In contrast to this, analytical expressions were used here.

## Stability limit

In accordance with the specified data sheet values the function Q(P) describes the reactive power control range in terms of P. However, for stability reasons the reactive power control range is further restricted. Natural stability is given for rotor angles between 0 and 90°. When considering an automatic voltage regulator (AVR) the natural stability region can be extended by an artificial stability region with load angles > 90 °. However, the influence of an AVR is for now not considered here. If a maximum angular load angle of 90° is assumed the following condition for minimal reactive power and stable operation can be formulated:

$Q = Q_{d} u \left(\frac{E_{p}}{U_{n}} \cos{\left (\delta \right )} - u\right)
~~\underrightarrow{\delta=90°}~~ 
Q_{min}= -u^2Q_d$

```python
QminMaxEqs =  sp.solve(eq3a+Psym,Qsym) 
#Solve P(Q)=0 resulting in two solutions for Qmin and Qmax
Qmin2 =-u**2*Qd
```

## Power chart plot

In order to plot the power chart, it is necessary to replace symbolic by numerical values.

```python
eq6a=eq3a.subs([(UnSym,Un),(QdSym,Qd),(EpSym,Ep),(UnSym,Un),(uSym,u)])
eq6b=eq3b.subs([(UnSym,Un),(QdSym,Qd),(EpSym,Ep),(UnSym,Un),(uSym,u)])
eq7a=eq5a.subs([(UnSym,Un),(QdSym,Qd),(EpSym,Ep),(UnSym,Un),(uSym,u)])
eq7b=eq5b.subs([(UnSym,Un),(QdSym,Qd),(EpSym,Ep),(UnSym,Un),(uSym,u)])

PaFun=sp.lambdify(Qsym,(eq6a+Psym),"numpy") 
PbFun=sp.lambdify(Qsym,(eq6b+Psym),"numpy") 
QaFun=sp.lambdify(Psym,(eq7a+Qsym),"numpy") 
QbFun=sp.lambdify(Psym,(eq7b+Qsym),"numpy") 
#http://docs.sympy.org/dev/tutorial/basic_operations.html

Qmin =float( QminMaxEqs[0].subs([(UnSym,Un),(QdSym,Qd),(EpSym,Ep),(UnSym,Un),
                                 (uSym,u)]) )
Qmax =float( QminMaxEqs[1].subs([(UnSym,Un),(QdSym,Qd),(EpSym,Ep),(UnSym,Un),
                                 (uSym,u)]) )
```

# Static stability

## Arrangement

Static stability of synchronous machines can be assessed with power angle characteristics.
An arrangement as shown in Figure 4a) is considered. A synchronous generator feeds its power through a transformer and a line into a medium-voltage network. The corresponding equivalent circuit of Figure 4b) can be further simplified to Figure 4c) which corresponds to the so called one-machine problem. 

<p></p><figure>
<img width=700  src="http://imgur.com/kko0zyN.png" alt='missing' />
</figure><p></p>
[Figure 4:](#Fig3) Considered arrangement and derived equivalent circuits.

According to <a name="ref-4"/>[(Crastan)](#cite-Crastan), 6.6.2.4 "Wirk und Blindleistungsabgabe" it is for the active and reactive power output of the round rotor generator:

$ P= \frac{E_{p} Q_{d}}{U_{r}} u \sin{\left (\delta \right )}$

In which $u$ is the per unit grid voltage $u=\frac{U}{U_r}$ and $\delta$ the load angle as $Q_d$ the so called characteristic reactive power:

$ Q_d=\frac{3U^2_r}{X_d}$

The synchronous d-component reactance $X_d$ of the generator is transformed by the square of the machine transformer ratio on the high voltage side:  $X_{dt}=a^2X_{d}$. Together with the resulting network impedance  $X_N=X_T+X_L+X_Q$ the characteristic reactive power an be defined to:  

$ Q_{dN}=\frac{3U^2_r}{X_{dt}+X_N}$

With $E_{pT}=aE_{p}$ the active power output of the arrangement according to Figure 1c) then results to:

$ P(\delta)= \frac{E_{pT} Q_{dN}}{U_{rt}} u \sin{\left (\delta \right )}$

*Note: The expression $E_{pt}$ is used in <a name="ref-4"/>[(Crastan)](#cite-Crastan) already elsewhere. For this reason $E_{pT}$ is used as symbol for the transformed internal voltage.*

The reactive power output of the arrangement according to Figure 1c) can be derived in an analogue way and is given to:

$ Q(\delta) = u Q_{dN} ( \frac{E_{pT}}{U_{rt}} \cos{\left(\delta \right )-u)}$

The equations for $P(\delta)$ and $Q(\delta)$ can be each solved for
$E_{pT}$. By further insertion the function $\delta(P,Q)$ is built.

```python
QdNsym,EpTsym,UrtSym=sp.symbols(r"Q_{dN},E_{pT},U_{rt}")
eq8= uSym*QdNsym*EpTsym/UrtSym  *sp.sin(dltSym) -Psym #P(dlt) term (Crst 6.94)
eq9= uSym*QdNsym*(EpTsym/UrtSym *sp.cos(dltSym) -uSym)-Qsym # Q(dlt) term 

eq10=sp.solve(eq9,EpTsym)[0] #Solve for EpT (one solution)
eq11=eq8.subs(EpTsym,eq10)
eq12=sp.solve(eq11,dltSym)[0]
#Solve for delta, gives two solutions for dlt(P,Q) function,
#Use only use first (stable one)solution.

display(sp.Eq(Psym,eq8))
display(sp.Eq(Qsym,eq9))
display(sp.Eq(dltSym,eq12))
```

## Power-angle plot

The $P(\delta)$ function describes as a sine function, the so-called power-angle characteristic.
The resulting plot for the arrangement according to [Figure 4](#Fig4) is given in [Figure 5](#Fig5). Operation points with positive derivative $dP/d\delta$ are stable whereas operation points with negative 
derivative (right of the maxima) are unstable.

```python
SnTf=SnTf_Widg.value
Urt=Urt_Widg.value
uk=uk_Widg.value
XsLine=XsLine_Widg.value

Ur=Un #Todo: Change Un to Ur
a= Urt/Ur
Xdt= Xd*a**2
X_T =uk/100* Urt**2/SnTf 

eq13=eq12.subs(EpTsym,Ep) 
dltFun=sp.lambdify((Psym,Qsym,uSym,QdNsym),eq13,"numpy") 
#Numerical dlt(P,Q,u,QdN) function
                                              
eq14=eq10.subs(UrtSym,Urt)
EpTfun=sp.lambdify((dltSym,Qsym,uSym,QdNsym),eq14,"numpy") 
#Numerical EpT(dlt,Q,u,QdN) function

def drawPowerAngleChar(P,cosPhi,u,LineLength,Skss):
  
   X_L=XsLine * LineLength 
   X_Q=Urt**2/Skss 
   X_N=X_T+X_L+X_Q
   QdN =3*Urt**2/(Xdt+X_N)
   Q= P*tan(arccos(cosPhi))
   
   dltDeg=dltFun(P,Q,u,QdN) *180/pi
   EpT    =EpTfun(dltDeg *pi/180,Q,u,QdN) 
   
   plt.plot(dltDeg,P,color="blue", marker='o',fillstyle='full',
            markeredgewidth=0.0)
 
   dltDegArr =np.linspace(0,180)  
   Parr =EpT*QdN/Urt*u*sin(dltDegArr *pi/180)   
   plt.plot(dltDegArr,Parr)
      
   plt.axhline(y=P,linewidth=1, linestyle = "--", color="red")
   plt.axvline(x=dltDeg,linewidth=1, linestyle = "--", color="red")
   
   plt.xlabel('Load angle $\delta$  [deg]')
   plt.ylabel('Active power $P$ [MW]')
   plt.grid(True)
   font = {'weight' : 'normal','size':18}
   plt.rc('font', **font) #Change the font size for all plot elements 
   
   #Add boxed text: http://matplotlib.org/users/recipes.html
   textStr=(
   '$Ep=%.4f $$\mathrm{kV}$$(%.2f $ $\mathrm{pu})$\n$\delta=%.2f^{\circ}$' 
   %(EpT,EpT/Ur,dltDeg) )
   
   props = dict(boxstyle='round', facecolor='white', alpha=0.5)
   plt.gca().text(0.05,0.95,textStr,transform=plt.gca().transAxes,fontsize=18,
   verticalalignment='top', bbox=props, color='black') 
 
   plt.ylim(0,20) #Comment for dynamic adjustment of axis scale
   plt.show()
  
P_Slider = widgets.FloatSlider(min=0,max=Pn,value=Pn,description="$P$ [MW] ")

cosPhi_Slider = widgets.FloatSlider(min=-1, max=1, 
                                   value= cosPhiN,description="$cos\\varphi$")
u_Slider = widgets.FloatSlider(min= 0, max= 1,
                             value=1,description="$u$ [pu]")
LineLength_Slider = widgets.FloatSlider(min=0, max=100,
                             value=0,description="$Line$ [km]")                                             
Skss_Slider = widgets.FloatSlider(min= 15, max=200,
                            value=round(100,3),description="$S_k{''}$ [MVA]")

widgets.interactive(lambda a,b,c,d,e: drawPowerAngleChar(a,b,c,d,e),
     a=P_Slider,b=cosPhi_Slider,c=u_Slider,d=LineLength_Slider,e=Skss_Slider)
```

# Synchronizing  power

The so-called synchronizing power of the synchronous machine is the derivative of the power angle for the load angle. It is a measure of the stability of the generator. Positive values indicate a stable operation, 0 is indifferent case and negative values indicate instability. According to <a name="ref-4"/>[(Crastan)](#cite-Crastan), 6.6.2.4 "Wirk und Blindleistungsabgabe" the synchronizing power is:
  
$\large P_{sync}= u Q_{d} \frac{E_{p}}{U_{r}} \cos{\left (\delta \right )}$

## Regular case plot

```python
@np.vectorize
def func(x,y): #X=P,Y=Skss,Z=Psync
    
    P=x
    Skss   =y
    
    X_L=XsLine * LineLength 
    X_Q=Urt**2/Skss 
    X_N=X_T+X_L+X_Q
    QdN =3*Urt**2/(Xdt+X_N)
    Q= P*tan(arccos(cosPhi))
   
    dltDeg=dltFun(P,Q,u,QdN) *180/pi
    EpT    =EpTfun(dltDeg *pi/180,Q,u,QdN) 
    
    Psync=u*QdN*EpT/Urt*cos(dltDeg *pi/180)    
    return Psync

cosPhi=cosPhi_Slider.value
u=u_Slider.value
LineLength =LineLength_Slider.value

N = 100
xReversed=np.linspace(-1*Pn,0,N,endpoint=False)
#Reversed and negative array (no startpoint=False option available)
x = -1*xReversed[::-1]
#x = np.linspace(0.0001, Pn, N)  #X=P
y = np.linspace(Skss_Slider.min,Skss_Slider.max, N) #Y=Skss,
X,Y = np.meshgrid(x,y)
X = X.T #needed?
Y = Y.T

Z = func(X, Y) #Z=Psync

CS=plt.contour(X, Y, Z,colors=('black'))
plt.clabel(CS, inline=1, fontsize=20)
plt.contourf(X, Y, Z,cmap='jet_r')
cbar = plt.colorbar(label='Synchronising power$P_{sync}$ [MW]')
plt.xlabel("Active power P [MVA]")
plt.ylabel("$Grid short circuit power S_k{''}$ [MVA]")
plt.grid(True)
font = {'weight' : 'normal','size':18}
plt.rc('font', **font)  #Change the font size for all plot elements 
plt.show()
```
